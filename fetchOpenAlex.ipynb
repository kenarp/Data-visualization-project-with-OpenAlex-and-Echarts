```python
import requests
import json 
import pandas as pd
import numpy as np
```

```python
# Get sub-concepts of HCI
URL = "https://api.openalex.org/concepts?per-page=200&filter=level:2|3|4|5,ancestors.id:C107457646"
res = requests.get(URL)
```

```python
data = json.loads(res.text)
print(type(data))
```

```python
for key in data:
    print (key)
```

```python
print(len(data["results"]))
concepts = data["results"]
```

```python
filtered_key = ['display_name','level','description','works_api_url']
```

```python
orgnized_concepts = [{key:val for key,val in a_concept.items() if key in filtered_key } for a_concept in concepts]
```

```python
orgnized_concepts
```

```python
concepts_url = {concept ['display_name']:{'api_url':concept ['works_api_url'],'concept_level':concept ['level'],'description':concept['description']} for concept in orgnized_concepts}
# concepts_url
```

```python
filterYears='2003'
for year in np.arange(2004,2023):
    filterYears += '|'+str(year)
filterYears
```

```python
worksCount = {}
for concept in concepts_url:
    url = concepts_url[concept]['api_url']
    filteredUrl = url +',publication_year:'+filterYears
    res = requests.get(filteredUrl)
    data=json.loads(res.text)
    concepts_url[concept]['counts'] = data['meta']['count']
#     print(concepts_url[concept])
```

```python
concepts_url
```

```python
df=pd.DataFrame(concepts_url).transpose()
df = df.reset_index(names=['concept'])
df
```

```python
df_sort_by_count = df.sort_values(by='counts', ascending=False,ignore_index=True)
df_sort_by_count.index = np.arange(1, len(df_sort_by_count) + 1)
df_sort_by_count.head(20)
```

```python
top12ConceptsDict = pd.Series(df_sort_by_count['api_url'].values[:12],index=df_sort_by_count['concept'][:12]).to_dict()

top12ConceptsDict
```

```python
with open('top12Concepts.json', 'w') as file:
    # write the dictionary to the file in JSON format
    json.dump(top12ConceptsDict, file)
```

```python
concepts_worksCount = {}
```

```python
for concept,url in top12ConceptsDict.items():
    concepts_worksCount[concept] = {}
    for year in np.arange(2003,2023).tolist():
        countGroupByYear = {}
        yearFilteredUrl = url +',publication_year:'+str(year)
        res0 = requests.get(yearFilteredUrl)
        data0=json.loads(res0.text)
        groupedUrl = yearFilteredUrl+'&group_by=authorships.institutions.type'
        countGroupByYear['total_works'] = data0['meta']['count']
        res1 = requests.get(groupedUrl)
        data1=json.loads(res1.text)
        for countByInstitution in data1['group_by']:
            if countByInstitution['key'] == 'education':
                countGroupByYear['education_works'] = countByInstitution['count']
            if countByInstitution['key'] == 'company':
                countGroupByYear['company_works'] =countByInstitution['count']
        concepts_worksCount[concept][year] = countGroupByYear
        print(concept+': '+str(year))
```

```python
concepts_worksCount
```

```python
with open('top12ConceptsWorksCount.json', 'w') as file:
    # write the dictionary to the file in JSON format
    json.dump(concepts_worksCount, file)
```
